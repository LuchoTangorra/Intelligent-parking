```python
def add_ocupation(d, inplace=False):
    '''
    Includes a new column called ocupacion and adds one to a row
        if its a "1" and substract one if its "0".
    
    Parameters:
        d: (DataFrame) all SUMO data
        inplace: (Boolean) if use a reference of d or make a copy instead.
        
    Return:
        out: (DataFrame) the d reference or copy that contains the new column ocupacion
            with the correct values.
    
    '''
    def sum_ocup(op):
        '''
        Just for the apply func. If the row value is Entrada we sum 1,
        otherwise we subs 1.
        On each day we restart the actual_ocup.
        '''
        nonlocal actual_ocup

        if op == 1:
            actual_ocup += 1
        else:
            if op == 0:
                if actual_ocup > 0:
                    actual_ocup -= 1
        return actual_ocup

    #To use a reference or a copy of the input
    if inplace:
        out = d
    else:
        out = d.copy()

    streets = d['parquimetro'].drop_duplicates().sort_values()
    days = d['fecha_emision'].drop_duplicates()
    
    out['ocupacion'] = 0
    for street in streets:
        clear_output()
        print(street)
        for day in days:
            print(day)
            actual_ocup = 0

            #actual_d = out.loc[(out.parquimetro == street)] #& (out.fecha_emision == day)]
            out.loc[(out.parquimetro == street) & (out.fecha_emision == day), 'ocupacion'] = out.loc[(out.parquimetro == street) & (out.fecha_emision == day), 'operacion'].apply(sum_ocup)

    if not inplace:
        return out
```

```python
def get_max_ocupation_streets(data):
    '''
    Obtain the max ocupacion per street.
    
    Parameters:
        data: (DataFrame) a DataFrame containing all the data.
        
    Returns:
        max_ocupation: (DataFrame) max ocupation for all streets.
    '''
    d = data.copy()
    max_ocupation = data[['parquimetro', 'direccion', 'ocupacion']].groupby(by='direccion', as_index=True).max().reset_index()
    return max_ocupation.sort_values(by='ocupacion', ascending=False)[['parquimetro','direccion','ocupacion']].reset_index(drop=True)

def get_mean_ocupation_streets(data):
    '''
    Obtain the mean ocupacion per street.
    
    Parameters:
        data: (DataFrame) a DataFrame containing all the data.
        
    Returns:
        mean_ocupation: (DataFrame) mean ocupation for all streets.
    '''
    d = data.copy()
    #mean_ocupation = d.groupby(by='parquimetro').mean()
    std_ocupation = data[['parquimetro', 'direccion', 'ocupacion']].groupby(by=['parquimetro', 'direccion'], as_index=True).std().reset_index()
    std_ocupation.rename(columns={'ocupacion': 'desvio_estandar'}, inplace=True)
    mean_ocupation = data[['parquimetro', 'direccion', 'ocupacion']].groupby(by=['parquimetro', 'direccion'], as_index=True).mean().reset_index()
    mean_ocupation = mean_ocupation.merge(std_ocupation[['parquimetro','desvio_estandar']], on='parquimetro')
    
    #error
    
    return mean_ocupation.sort_values(by='ocupacion', ascending=False)[['parquimetro','direccion','ocupacion', 'desvio_estandar']].reset_index(drop=True)
```

```python
def traffic_density(data):
    '''
    Obtain the traffic density (how much in-outs) per street.
    
    Parameters:
        data: (DataFrame) a DataFrame containing all the data.
        
    Returns:
        d: (DataFrame) all traffic density for all streets.
    '''
    d = data.copy()
    d = d[['parquimetro','direccion', 'ocupacion']].groupby(by=['parquimetro','direccion'], as_index=True).count().reset_index()
    d.rename(columns={'ocupacion': 'movimientos'}, inplace=True)
    d = d.sort_values(by='movimientos')
    return d.sort_values(by='movimientos', ascending=False).reset_index(drop=True)
```

```python
def plot_mean_parked_cars(data, street):
    '''
        Plot a line graph for the mean parked cars along a street.
        
        Parameters:
            data: (DataFrame) a DataFrame containing all the data.
            street: (int) the street to chose the data.
    '''
    
    d = data.copy()
    d = d.loc[d['parquimetro'] == street]
    mean_value = d['ocupacion'].mean()
    std_value = d['ocupacion'].std()
    
    std = d[['tiempo_emision', 'ocupacion']].groupby(by='tiempo_emision').std().reset_index()
    d = d[['tiempo_emision', 'ocupacion']].groupby(by='tiempo_emision').mean().reset_index()
    
    d.rename(columns={'ocupacion': 'mean_ocupacion'}, inplace=True)
    std.rename(columns={'ocupacion': 'std_ocupacion'}, inplace=True)
    
    #std.fillna(0, inplace=True)
    
    std_1 = d['mean_ocupacion'] + std['std_ocupacion']
    std_2 = d['mean_ocupacion'] - std['std_ocupacion']
                
    #Modify the plot size
    plt.rcParams['figure.figsize'] = (25,14)
    
    #Set the label titles
    plt.xlabel('Time')
    plt.ylabel('Parked cars')
    plt.title('Parked cars in street nº{}. Mean: {}. Std: {}. Sample of {} parked cars'.format(street, mean_value, std_value, len(d)))
    
    #Set the plot data for each graph (ax1,ax2,ax3)
    plt.plot(d['tiempo_emision'], d['mean_ocupacion'], color='black')
    plt.fill_between(d['tiempo_emision'].to_numpy(), std_1.to_numpy(), std_2.to_numpy(), alpha='0.3', color='grey')
    
    #plt.legend()
    plt.show()
```

```python
def inputs_outputs_per_hour(data):
    '''
    Obtain the inputs and outputs (and the sub of both) per street.
    
    Parameters:
        SUMO_data: (DataFrame) a DataFrame containing all the data or only for
        one day, e.g. only with the monday data.
        
    Returns:
        in_out_per_hour: (DataFrame) df that contains the inputs, the outputs 
        and the sub: entradas-salidas.
    '''
    
    d = data.copy()
    d = d[['parquimetro','tiempo_emision','operacion']]
    
    d['tiempo_emision'] = d['tiempo_emision'].apply(lambda x: x.hour)

    #Obtain 2 df, one with the count of the inputs and the other with the count of the outputs
    in_per_hour = d.loc[d['operacion'] == 1, ['tiempo_emision','parquimetro', 'operacion']].groupby(by=['tiempo_emision','parquimetro']).count().reset_index()
    out_per_hour = d.loc[d['operacion'] == 0, ['tiempo_emision','parquimetro', 'operacion']].groupby(by=['tiempo_emision','parquimetro']).count().reset_index()
    in_per_hour.rename(columns={'operacion': 'entradas'}, inplace=True)
    out_per_hour.rename(columns={'operacion': 'salidas'}, inplace=True)

    in_per_hour = in_per_hour.groupby(by=['parquimetro','tiempo_emision']).mean().reset_index()
    out_per_hour = out_per_hour.groupby(by=['parquimetro','tiempo_emision']).mean().reset_index()
    
    #Merge both in and out per hour to get the diff
    in_out_per_hour = in_per_hour.merge(out_per_hour, how='outer', on=['parquimetro','tiempo_emision'])
    in_out_per_hour.fillna(0, inplace=True)

    in_out_per_hour['entrada - salida'] = in_out_per_hour['entradas'] - in_out_per_hour['salidas']
    in_out_per_hour.rename(columns={'tiempo': 'hora'}, inplace=True)
    
    return in_out_per_hour
```

```python
in_outs_per_hour = inputs_outputs_per_hour(data)

#Parkimeter to show its in_outs and if order by time or by in-outs
par = 1
by_time = True

in_outs_per_hour = in_outs_per_hour.loc[in_outs_per_hour['parquimetro'] == par].sort_values(by='tiempo_emision').reset_index(drop=True)

if by_time:
    display(in_outs_per_hour.sort_values(by='tiempo_emision'))
else:
    display(in_outs_per_hour.sort_values(by='entrada - salida'))
```

```python
def get_data_to_plot(data, hour):
    '''
    
    '''
    d = data.copy()
    d = d[['parquimetro','tiempo_emision', 'fecha_emision', 'direccion', 'ocupacion']]
    d['tiempo_emision'] = d['tiempo_emision'].apply(lambda x: x.hour)
    d = d.loc[d['tiempo_emision'] == hour]
    d = d.groupby(by=['parquimetro', 'direccion']).mean().reset_index()
    return d
    
    
def get_color_to_plot(data, max_ocup, colors=['green','yellow','orange','red']):
    '''
    
    '''
    d = data.copy()
    #max_ocup = get_max_ocupation_streets(d)
    
    d = d.sort_values(by='parquimetro').reset_index(drop=True)
    #max_ocup = max_ocup.sort_values(by='parquimetro').reset_index(drop=True)
    
    #print(d)
    #print(max_ocup)
    
    grade_1 = d['ocupacion'] <= (max_ocup['ocupacion'] * 1/4)
    grade_2 = (~grade_1) & (d['ocupacion'] <= (max_ocup['ocupacion'] * 2/4))
    grade_3 = (~grade_1) & (~grade_2) & (d['ocupacion'] <= (max_ocup['ocupacion'] * 3/4))
    grade_4 = (~grade_1) & (~grade_2) & (~grade_3)

    color_data = pd.DataFrame(index=d.index.values, columns=['color'])
    color_data[grade_1] = colors[0]
    color_data[grade_2] = colors[1]
    color_data[grade_3] = colors[2]
    color_data[grade_4] = colors[3]

    color_list = []
    for c in color_data.values:
        color_list.append(c[0])

    color_data = color_list
    
    return color_data

def add_latitud_longitud(data, lat_lon, inplace=False):
    '''
    
    '''
    if inplace:
        out = data
    else:
        out = data.copy()
    
    lat_lon = lat_lon[['parquimetro','latitud','longitud']].drop_duplicates(subset='parquimetro')
    out = out.merge(lat_lon[['parquimetro','latitud','longitud']], on='parquimetro')
    if not inplace:
        return out
    
def plot_scatter_map(data, colour, date=None):
    plt.scatter(data['latitud'], data['longitud'], c=colour, s=100)
    plt.title('Scatter for hour {}'.format(hour))
    plt.axis('off')
    plt.show()
```
